Import OpenCV library.

Read the input image.

Apply different filters:

Averaging filter

Gaussian filter

Median filter

Bilateral filter

Display all filtered images.

```python
import cv2
import matplotlib.pyplot as plt

# -------- 1. Import & Read Image --------
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg")
img = cv2.resize(img, (500, 500))   # resize for clarity
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------- 2. Apply Filters --------

# Averaging Filter
avg = cv2.blur(img_rgb, (5,5))

# Gaussian Filter
gaussian = cv2.GaussianBlur(img_rgb, (5,5), 0)

# Median Filter
median = cv2.medianBlur(img_rgb, 5)

# Bilateral Filter
bilateral = cv2.bilateralFilter(img_rgb, 9, 75, 75)

# -------- 3. Display --------
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Averaging")
plt.imshow(avg)
plt.axis('off')

plt.subplot(2,3,3)
plt.title("Gaussian")
plt.imshow(gaussian)
plt.axis('off')

plt.subplot(2,3,4)
plt.title("Median")
plt.imshow(median)
plt.axis('off')

plt.subplot(2,3,5)
plt.title("Bilateral")
plt.imshow(bilateral)
plt.axis('off')

plt.tight_layout()
plt.show()
```
