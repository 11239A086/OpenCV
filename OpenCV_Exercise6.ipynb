Read the image in grayscale.
Choose thresholding methods:
Apply
1. Simple Thresholding
2. Adaptive Thresholding
3. Otsu’s Thresholding
4. the thresholding function.

Display the binary images.
Analyze the segmentation results.

```python
import cv2
import matplotlib.pyplot as plt

# -------- 1. Read Image in Grayscale --------
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg", 0)  # 0 = grayscale
img = cv2.resize(img, (500, 500))

# -------- 2. Apply Thresholding --------

# Simple Threshold
_, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

# Adaptive Threshold
th2 = cv2.adaptiveThreshold(img, 255,
                            cv2.ADAPTIVE_THRESH_MEAN_C,
                            cv2.THRESH_BINARY,
                            11, 2)

# Otsu Threshold
_, th3 = cv2.threshold(img, 0, 255,
                       cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# -------- 3. Display --------
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title("Original (Gray)")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Simple Threshold")
plt.imshow(th1, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Adaptive Threshold")
plt.imshow(th2, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Otsu Threshold")
plt.imshow(th3, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
