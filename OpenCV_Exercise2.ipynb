Display image properties (shape, size, datatype).

Color Space Conversion (RGB, Grayscale, HSV)

Split and merge color channels.

Extract Region of Interest (ROI).

Change image resolution and aspect ratio.

```python
import cv2
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg")

# Convert BGR → RGB (important for matplotlib)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------- 1. Image Properties --------
print("Shape:", img.shape)
print("Size:", img.size)
print("Datatype:", img.dtype)

# -------- 2. Color Space Conversion --------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# -------- 3. Split & Merge Channels --------
b, g, r = cv2.split(img)
merged = cv2.merge([b, g, r])
merged_rgb = cv2.cvtColor(merged, cv2.COLOR_BGR2RGB)

# -------- 4. ROI --------
roi = img_rgb[50:200, 50:200]

# -------- 5. Resize --------
resized = cv2.resize(img_rgb, (300, 300))
aspect = cv2.resize(img_rgb, None, fx=0.5, fy=0.5)

# -------- Display Everything --------
plt.figure(figsize=(12,10))

plt.subplot(3,3,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(3,3,2)
plt.title("Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.subplot(3,3,3)
plt.title("HSV")
plt.imshow(hsv)
plt.axis('off')

plt.subplot(3,3,4)
plt.title("Blue Channel")
plt.imshow(b, cmap='gray')
plt.axis('off')

plt.subplot(3,3,5)
plt.title("Green Channel")
plt.imshow(g, cmap='gray')
plt.axis('off')

plt.subplot(3,3,6)
plt.title("Red Channel")
plt.imshow(r, cmap='gray')
plt.axis('off')

plt.subplot(3,3,7)
plt.title("ROI")
plt.imshow(roi)
plt.axis('off')

plt.subplot(3,3,8)
plt.title("Resized")
plt.imshow(resized)
plt.axis('off')

plt.subplot(3,3,9)
plt.title("Aspect Ratio")
plt.imshow(aspect)
plt.axis('off')

plt.tight_layout()
plt.show()
```
