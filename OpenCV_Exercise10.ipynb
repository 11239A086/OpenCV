Read the input image.

Convert the image to grayscale.

Apply histogram calculation on the grayscale image (single intensity channel).

Apply histogram calculation on individual color channels (B, G, R). Display the original image, grayscale image, and all histograms.

Compare the outputs visually.

```python
import cv2
import matplotlib.pyplot as plt

# -------- 1. Read Image --------
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg")
img = cv2.resize(img, (500, 500))

# -------- 2. Convert to RGB --------
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------- 3. Convert to Grayscale --------
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# -------- 4. Plot Histograms --------
plt.figure(figsize=(12,6))

# Original Image
plt.subplot(2,3,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

# Grayscale Image
plt.subplot(2,3,2)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis('off')

# Grayscale Histogram (Single Channel)
plt.subplot(2,3,3)
plt.title("Gray Histogram")
plt.hist(gray.ravel(), bins=256)
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# -------- 5. RGB Histograms --------
colors = ('r', 'g', 'b')

plt.subplot(2,1,2)
plt.title("RGB Histograms")

for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(hist, color=col)

plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
