Morphological Operations on Dataset Import required libraries (cv2, os, numpy).

Load all images from dataset folder.

Convert each image to grayscale.

Apply thresholding (to make it binary).

Apply morphological operations:

Erosion

Dilation

Opening

Closing

Gradient

Display or save the processed images.

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Dataset path
folder = '/content/drive/MyDrive/dataset'

# Kernel for operations
kernel = np.ones((5,5), np.uint8)

# Loop through all images
for file in os.listdir(folder):
    path = os.path.join(folder, file)

    img = cv2.imread(path)
    if img is None:
        continue

    # Grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Threshold (Binary)
    _, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # Morphological Operations
    erosion = cv2.erode(thresh, kernel, iterations=1)
    dilation = cv2.dilate(thresh, kernel, iterations=1)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
    closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
    gradient = cv2.morphologyEx(thresh, cv2.MORPH_GRADIENT, kernel)

    # Display using matplotlib
    titles = ['Original', 'Gray', 'Binary', 'Erosion', 'Dilation', 'Opening', 'Closing', 'Gradient']
    images = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB), gray, thresh, erosion, dilation, opening, closing, gradient]

    plt.figure(figsize=(12,6))
    for i in range(len(images)):
        plt.subplot(2,4,i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.show()
```
