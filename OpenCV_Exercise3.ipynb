Resize an image using scaling factors.

Rotate an image by a given angle.

Crop a specific portion of an image.

Flip image horizontally and vertically.

Adjust brightness and contrast.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load image
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# -------- 1. Resize using Scaling Factors --------
scaled = cv2.resize(img_rgb, None, fx=0.5, fy=0.5)

# -------- 2. Rotate Image --------
(h, w) = img.shape[:2]
center = (w//2, h//2)
matrix = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 degree
rotated = cv2.warpAffine(img_rgb, matrix, (w, h))

# -------- 3. Crop Image --------
crop = img_rgb[150:3000, 150:3000]

# -------- 4. Flip Image --------
flip_h = cv2.flip(img_rgb, 1)   # horizontal
flip_v = cv2.flip(img_rgb, 0)   # vertical

# -------- 5. Brightness & Contrast --------
bright = cv2.convertScaleAbs(img_rgb, alpha=1, beta=50)   # brightness ↑
contrast = cv2.convertScaleAbs(img_rgb, alpha=2, beta=0)  # contrast ↑

# -------- Display --------
plt.figure(figsize=(12,10))

plt.subplot(3,3,1)
plt.title("Original")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(3,3,2)
plt.title("Scaled (0.5x)")
plt.imshow(scaled)
plt.axis('off')

plt.subplot(3,3,3)
plt.title("Rotated 45°")
plt.imshow(rotated)
plt.axis('off')

plt.subplot(3,3,4)
plt.title("Cropped")
plt.imshow(crop)
plt.axis('off')

plt.subplot(3,3,5)
plt.title("Flip Horizontal")
plt.imshow(flip_h)
plt.axis('off')

plt.subplot(3,3,6)
plt.title("Flip Vertical")
plt.imshow(flip_v)
plt.axis('off')

plt.subplot(3,3,7)
plt.title("Brightness +")
plt.imshow(bright)
plt.axis('off')

plt.subplot(3,3,8)
plt.title("Contrast +")
plt.imshow(contrast)
plt.axis('off')

plt.tight_layout()
plt.show()
```
