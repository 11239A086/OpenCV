Write a Program to compare Sobel, Laplacian, and Canny edge detection methods.

Read image in grayscale.

Apply Sobel (X and Y). Apply Laplacian. Apply Canny.

Display all outputs.

Write the difference between all three.

```python
import cv2
import matplotlib.pyplot as plt

# -------- 1. Read Image in Grayscale --------
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg", 0)
img = cv2.resize(img, (500, 500))

# -------- 2. Sobel Edge Detection --------
sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)

# -------- 3. Laplacian --------
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# -------- 4. Canny --------
canny = cv2.Canny(img, 100, 200)

# -------- 5. Display --------
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.title("Original")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,3,2)
plt.title("Sobel X")
plt.imshow(sobelx, cmap='gray')
plt.axis('off')

plt.subplot(2,3,3)
plt.title("Sobel Y")
plt.imshow(sobely, cmap='gray')
plt.axis('off')

plt.subplot(2,3,4)
plt.title("Laplacian")
plt.imshow(laplacian, cmap='gray')
plt.axis('off')

plt.subplot(2,3,5)
plt.title("Canny")
plt.imshow(canny, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
