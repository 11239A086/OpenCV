Draw a line on an image.

Draw a rectangle and circle on an image.

Draw multiple geometric shapes.

Write text on an image.

Create a simple drawing canvas using OpenCV.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------- Load & Resize Image (IMPORTANT) --------
img = cv2.imread("/content/drive/MyDrive/dataset/img.jpg")
img = cv2.resize(img, (500, 500))   # make size fixed
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

draw = img.copy()

# -------- 1. Draw Line --------
cv2.line(draw, (50, 250), (450, 250), (255, 0, 0), 8)

# -------- 2. Rectangle --------
cv2.rectangle(draw, (50, 50), (200, 200), (0, 255, 0), 8)

# -------- 3. Circle --------
cv2.circle(draw, (350, 150), 80, (0, 0, 255), 8)

# -------- 4. Multiple Shapes --------
cv2.line(draw, (0, 0), (200, 200), (255, 255, 0), 4)
cv2.circle(draw, (250, 350), 50, (0, 255, 255), -1)  # filled

# -------- 5. Text --------
cv2.putText(draw, "OpenCV", (120, 480),
            cv2.FONT_HERSHEY_SIMPLEX,
            1.5, (255, 255, 255), 4)

# -------- Canvas (No Image Needed) --------
canvas = np.ones((400, 400, 3), dtype="uint8") * 255

cv2.line(canvas, (50, 50), (350, 50), (255, 0, 0), 5)
cv2.rectangle(canvas, (50, 100), (200, 300), (0, 255, 0), 5)
cv2.circle(canvas, (300, 200), 60, (0, 0, 255), -1)

cv2.putText(canvas, "Canvas", (100, 370),
            cv2.FONT_HERSHEY_SIMPLEX,
            1, (0, 0, 0), 2)

canvas = cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB)

# -------- Display --------
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Image with Shapes")
plt.imshow(draw)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Drawing Canvas")
plt.imshow(canvas)
plt.axis('off')

plt.tight_layout()
plt.show()
```
